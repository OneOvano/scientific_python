```python
x = np.r_[0:10:100j]  # np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-((x - 1)**2 / 8))
plt.xlabel(r'$\nabla$')
plt.ylabel(r'$y$')
plt.title('Figure')
plt.plot(x, y1, '-', label='sin')
plt.plot(x, y2, '--', label='cos', color=(0, 0.8, 1))
plt.plot(x, y3, ':', label='Gauss', color='#ff0000')
plt.legend()
plt.grid()

plt.figure()
plt.plot(np.arange(10)**2)
```
