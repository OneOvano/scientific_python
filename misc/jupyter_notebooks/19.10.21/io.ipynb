```python
fname = './diagram.csv'
x = []
y = []
with open(fname) as f:
    for row in f:
        row = row.strip()
        x_, y_ = map(float, row.split(','))
        x.append(x_)
        y.append(y_)
# print(x)
# print(y)

basename, ext = os.path.splitext(fname)
# print(basename, ext)
dirname = './data'
os.makedirs(dirname, exist_ok=True)

new_fname = os.path.join(dirname, basename + '.tsv')

with open(new_fname, 'w') as f:
    for x_, y_ in zip(x, y):
        f.write('{x}\t{y}\n'.format(x=x_, y=y_))
```

```python
def convert_magn(s):
    if s.startswith('<'):
        s = s[1:]
    return float(s)

df['m'] = df.Magnitude.map(convert_magn)
df['is_upper_limit'] = df.Magnitude.map(lambda s: s.startswith('<'))
df.is_upper_limit
```

```python
plt.plot(df.JD[~df.is_upper_limit], df.m[~df.is_upper_limit], 'x')
plt.plot(df.JD[df.is_upper_limit], df.m[df.is_upper_limit], 'v')
plt.ylim(*reversed(plt.ylim()))
```
